```python
import requests
from bs4 import BeautifulSoup


# Make a GET request to the website
url = "https://books.toscrape.com/"
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

# Find all the categories
categories = soup.find_all("div", {"class": "side_categories"} )

# Loop through each category and find all the books
for category in categories:
    books_url = category.find("a")["href"]
    full_books_url = url + books_url

    # Make a GET request to the books page
    books_response = requests.get(full_books_url)
    books_soup = BeautifulSoup(books_response.content, "html.parser")

    # Find all the books on the page
    books = books_soup.find_all("article", {"class": "product_pod"})

    # Loop through each book and print its details
    for book in books:
        # Extract the book's title, author, price, and availability
        title = book.find("h3").find("a")["title"]
        price = book.find("p", {"class": "price_color"}).text.strip()
        availability = book.find("p", {"class": "availability"}).text.strip()

        # Extract the book's cover image URL and download it
        image_url = book.find("div", {"class": "image_container"}).find("img")["src"]
        image_response = requests.get(url + image_url)
        with open(f"{title}.jpg", "wb") as f:
            f.write(image_response.content)

        try:
            # Extract the book's description and additional details
            product_url = book.find("h3").find("a")["href"]
            full_product_url = url + product_url
            product_response = requests.get(full_product_url)
            product_soup = BeautifulSoup(product_response.content, "html.parser")
            product_information = product_soup.find("table", {"class": "table-striped"}).find_all("tr")

            # Extract the book's product code, category, and rating
            product_code = product_information[0].find("td").text.strip()
            category = product_information[1].find("td").text.strip()
            rating = product_soup.find("p", {"class": "star-rating"})["class"][1]

            # Extract the book's description
            description = product_soup.find("article", {"class": "product_page"}).find("p", recursive=False).text.strip()

            # Print the book's details
            print("Title: ", title)
            print("Price: ", price)
            print("Availability: ", availability)
            print("Product code: ", product_code)
            print("Category: ", category)
            print("Rating: ", rating)
            print("Description: ", description)
            print("Image URL: ", url + image_url)
            print("----------------------------------------------------")
        except AttributeError:
            print(f"Skipping book: {title} - information not found")
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website
url = "https://books.toscrape.com"

# Send a request to the website and get the HTML response
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Find all the categories in the side menu
categories = soup.find("div", {"class": "side_categories"}).find_all("a")

# Loop through each category and get the products
for category in categories:
    if category.text.strip() != "Books":
        category_url = url + "/" + category["href"]
        category_response = requests.get(category_url)
        category_soup = BeautifulSoup(category_response.content, "html.parser")
        products = category_soup.find_all("article", {"class": "product_pod"})
        for product in products:
            title = product.find("h3").find("a")["title"]
            product_url = url + "/" + product.find("a")["href"]
            product_response = requests.get(product_url)
            product_soup = BeautifulSoup(product_response.content, "html.parser")
            product_information = product_soup.find("table", {"class": "table-striped"})
            if product_information:
                product_information = product_information.find_all("tr")
                product_code = product_information[0].find("td").text.strip()
                category = product_information[1].find("td").text.strip()
                rating = product_soup.find("p", {"class": "star-rating"})["class"][1]
            else:
                product_code = "N/A"
                category = "N/A"
                rating = "N/A"
            author = product_soup.find("div", {"class": "product_main"}).find_all("a")[0].text.strip()
            price = product_soup.find("div", {"class": "product_main"}).find("p", {"class": "price_color"}).text.strip()
            availability = product_soup.find("div", {"class": "product_main"}).find("p", {"class": "availability"}).text.strip()
            image_url = url + "/" + product_soup.find("div", {"class": "item"}).find("img")["src"]
            description = product_soup.find("div", {"id": "product_description"}).find_next("p").text.strip()
            print("Title:", title)
            print("Author:", author)
            print("Price:", price)
            print("Availability:", availability)
            print("Product code:", product_code)
            print("Category:", category)
            print("Rating:", rating)
            print("Image URL:", image_url)
            print("Description:", description)
            print("\n")
```

```python
import requests
from bs4 import BeautifulSoup

base_url = "https://books.toscrape.com/"

# Send a GET request to the base URL and create a BeautifulSoup object
response = requests.get(base_url)
soup = BeautifulSoup(response.content, "html.parser")

# Find the "Books" category and get all the subcategories
books_category = soup.find("a", text="Books")
subcategories = books_category.find_next("ul").find_all("a")

# Loop through each subcategory
for subcategory in subcategories:
    subcategory_url = base_url + subcategory["href"]
    subcategory_response = requests.get(subcategory_url)
    subcategory_soup = BeautifulSoup(subcategory_response.content, "html.parser")

    # Find all the books on the current subcategory page
    products = subcategory_soup.find_all("article", {"class": "product_pod"})

    # Loop through each product and print the title, author, price, availability, image url, description, product code, category, and rating
    for product in products:
        title = product.find("h3").find("a")["title"]
        image_url = base_url + product.find("div", {"class": "image_container"}).find("img")["src"]
        product_url = base_url + product.find("div", {"class": "image_container"}).find("a")["href"]
        product_response = requests.get(product_url)
        product_soup = BeautifulSoup(product_response.content, "html.parser")
        
        try:
            author = product_soup.find("div", {"class": "product_main"}).find_all("a")[0].text.strip()
        except:
            author = "N/A"
        
        price = product_soup.find("div", {"class": "product_main"}).find("p", {"class": "price_color"}).text.strip()
        availability = product_soup.find("div", {"class": "product_main"}).find("p", {"class": "availability"}).text.strip()
        description = product_soup.find("div", {"id": "product_description"}).find_next("p").text.strip()
        
        product_information = product_soup.find("table", {"class": "table-striped"}).find_all("tr")
        product_code = product_information[0].find("td").text.strip()
        category = product_information[1].find("td").text.strip()
        rating = product_soup.find("div", {"class": "col-sm-6 product_main"}).find_all("p")[2]['class'][1]
        
        print("Title:", title)
        print("Author:", author)
        print("Price:", price)
        print("Availability:", availability)
        print("Image URL:", image_url)
        print("Description:", description)
        print("Product Code:", product_code)
        print("Category:", category)
        print("Rating:", rating)
        print("--------------------------------------------------")
```

```python
import requests
from bs4 import BeautifulSoup

# Make a GET request to the website
url = "https://books.toscrape.com/"
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

# Find all the links to individual book pages
links = soup.find_all("a", href=True)

# Loop through each book link and scrape the product description and information
for link in links:
    # Check if the link is to a book page
    if "catalogue/" in link["href"]:
        # Make a GET request to the book page
        book_url = url + link["href"]
        book_response = requests.get(book_url)

        # Parse the HTML content using Beautiful Soup
        book_soup = BeautifulSoup(book_response.content, "html.parser")

        # Scrape the product description and information
        title = book_soup.find("h1").text.strip()
        
        # Check if the title is 'A Light in the Attic'
        if title == 'A Light in the Attic':
            description_element = book_soup.find("div", {"id": "product_description"})
            description = description_element.find_next("p").text.strip() if description_element else ""
            
            table = book_soup.find("table", {"class": "table table-striped"})
            rows = table.find_all("tr") if table else []
            upc = rows[0].find("td").text.strip() if rows else ""
            price = rows[2].find("td").text.strip() if rows else ""
            availability = rows[5].find("td").text.strip() if rows else ""

            # Print the results
            print("Title:", title)
            print("Description:", description)
            print("UPC:", upc)
            print("Price:", price)
            print("Availability:", availability)
            print("-------------------------------------------------")
```

```python
import requests
from bs4 import BeautifulSoup

# Define the base URL for the website
base_url = "https://books.toscrape.com/catalogue/category/books_1/page-{}.html"

# Loop through each page of the website
for page in range(1, 51):
    # Make a GET request to the page
    response = requests.get(base_url.format(page))

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all the links to individual book pages
    links = soup.find_all("a", href=True)

    # Loop through each book link and scrape the product description and information
    for link in links:
        # Check if the link is to a book page
        if "catalogue/" in link["href"]:
            # Make a GET request to the book page
            book_url = "https://books.toscrape.com/" + link["href"]
            book_response = requests.get(book_url)

            # Parse the HTML content using Beautiful Soup
            book_soup = BeautifulSoup(book_response.content, "html.parser")

            # Scrape the product description and information
            title = book_soup.find("h1").text.strip()
            description_element = book_soup.find("div", {"id": "product_description"})
            description = description_element.find_next("p").text.strip() if description_element else ""

            table = book_soup.find("table", {"class": "table table-striped"})
            rows = table.find_all("tr") if table else []
            upc = rows[0].find("td").text.strip() if rows else ""
            price = rows[2].find("td").text.strip() if rows else ""
            availability = rows[5].find("td").text.strip() if rows else ""

            # Print the results
            print("Title:", title)
            print("Description:", description)
            print("UPC:", upc)
            print("Price:", price)
            print("Availability:", availability)
            print("-------------------------------------------------")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Make a GET request to the website
url = "https://books.toscrape.com/"
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

# Find all the links to individual book pages
links = soup.find_all("a", href=True)

# Create an empty list to store the scraped data
data = []

# Loop through each book link and scrape the product description and information
for link in links:
    # Check if the link is to a book page
    if "catalogue/" in link["href"]:
        # Make a GET request to the book page
        book_url = url + link["href"]
        book_response = requests.get(book_url)

        # Parse the HTML content using Beautiful Soup
        book_soup = BeautifulSoup(book_response.content, "html.parser")

        # Scrape the product description and information
        title = book_soup.find("h1").text.strip()
        
        description_element = book_soup.find("div", {"id": "product_description"})
        description = description_element.find_next("p").text.strip() if description_element else ""
        
        table = book_soup.find("table", {"class": "table table-striped"})
        rows = table.find_all("tr") if table else []
        upc = rows[0].find("td").text.strip() if rows else ""
        price = rows[2].find("td").text.strip() if rows else ""
        availability = rows[5].find("td").text.strip() if rows else ""

        # Store the data in a dictionary and add it to the list
        book_data = {
            "Title": title,
            "Description": description,
            "UPC": upc,
            "Price": price,
            "Availability": availability
        }
        data.append(book_data)

# Create a Pandas dataframe from the list of dictionaries
df = pd.DataFrame(data)

# Save the dataframe to a CSV file
df.to_csv("books.csv", index=False)

# Print the first few rows of the dataframe
print(df.head())
```

```python
import requests
from bs4 import BeautifulSoup

# Make a GET request to the website
url = "https://books.toscrape.com/"
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

# Find the number of pages
num_pages = soup.find("li", {"class": "current"}).text.split()[-1]

# Find the number of categories
category_list = soup.find("ul", {"class": "nav-list"}).find_all("li")
num_categories = len(category_list)

# Find the number of books per page
num_books_per_page = len(soup.find_all("article"))

# Print the results
print("Number of pages:", num_pages)
print("Number of categories:", num_categories)
print("Number of books per page:", num_books_per_page)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://books.toscrape.com/'

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

book_list = soup.find_all('article', {'class': 'product_pod'})

for book in book_list:
    title = book.select('h3 a')[0]['title']
    if title == 'A Light in the Attic':
        desc = book.p.text
        td = book.find('td')
        if td:
            upc = td.text
        else:
            upc = ''
        price = book.select('div p.price_color')[0].text
        avail = book.select('div p.availability')[0].text.strip()
        
        print('Title:', title)
        print('Description:', desc)
        print('UPC:', upc)
        print('Price:', price)
        print('Availability:', avail)
        print('-------------------------------------------------')

```
